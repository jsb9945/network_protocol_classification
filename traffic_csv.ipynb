```python
import os
from scapy.all import *
import csv
from scapy.layers.inet import TCP

# 설정
pcap_dir = "pcap/FTP/"  # 분석할 PCAP 파일들이 있는 디렉터리 경로
output_csv = "payload_csv/ftp_payloads.csv"  # FTP 패킷을 저장할 CSV 파일 경로

# CSV 파일 생성 및 헤더 작성
with open(output_csv, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['PCAP File', 'Payload(Bytes)'])  # CSV 파일 헤더

    # 디렉터리 내 모든 PCAP 파일 순회
    for pcap_file in os.listdir(pcap_dir):
        if pcap_file.endswith(".pcap"):  # PCAP 파일만 처리
            packets = rdpcap(os.path.join(pcap_dir, pcap_file))  # PCAP 파일 읽기

            # FTP 패킷 필터링 및 추출
            for i, packet in enumerate(packets):
                if packet.haslayer(TCP):
                    payload = bytes(packet[TCP].payload)
                    if payload:
                        writer.writerow([i+1, payload])
        print("finish")
```

```python
import os
from scapy.all import *
import csv
from scapy.layers.inet import TCP

# 설정
pcap_dir = "pcap/DNS/"  # 분석할 PCAP 파일들이 있는 디렉터리 경로
output_csv = "payload_csv/dns_payloads.csv"  # DNS 패킷을 저장할 CSV 파일 경로

# CSV 파일 생성 및 헤더 작성
with open(output_csv, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['PCAP File', 'Payload(Bytes)'])  # CSV 파일 헤더
    j=0
    # 디렉터리 내 모든 PCAP 파일 순회
    for pcap_file in os.listdir(pcap_dir):
        if pcap_file.endswith(".pcap"):  # PCAP 파일만 처리
            packets = rdpcap(os.path.join(pcap_dir, pcap_file))  # PCAP 파일 읽기
            
            # DNS 패킷 필터링 및 추출
            for i, packet in enumerate(packets):
                if packet.haslayer(DNS):
                    dns_layer = packet[DNS]
                    payload = bytes(dns_layer)  # DNS 패킷의 페이로드를 바이트로 추출
                    writer.writerow([j+1, payload])  # 페이로드만 저장
                    j+=1
        print("finish")
```
